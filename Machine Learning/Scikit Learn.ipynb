# Linear Regression

```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
df=pd.read_csv("income.data.csv")
print(df.head())
```

```python
#plotting the dataset by using scatter plot
plt.scatter(df.income,df.happiness,color="red",marker="*")
plt.xlabel("Income")
plt.ylabel("Happiness")
```

# Defining X and Y

```python
X=df["income"].values.reshape(-1,1)
y=df["happiness"].values.reshape(-1,1)
print(X)
```

# Splitting the dataset into the training set and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
print(X_train)
```

# Importing model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
#Fitting the model
lr.fit(X_train,y_train)
#prdict
y_pred=lr.predict(X_test)
print("Predicted values :",y_pred)
#printing values
print("Value of Intercept :",lr.intercept_)
print("Value of Coefficient :",lr.coef_)
```

# Model Evaluation

```python
from sklearn import metrics
```

```python
rmse=np.sqrt(metrics.mean_squared_error(y_test,y_pred))
R2=metrics.r2_score(y_test,y_pred) #r2_score always between 0 to 1
print("RMSE Score :",rmse)
print("R2 Score :",R2)
```

# Comparison of actual values and predicted values

```python
dt=pd.DataFrame({"Actual":y_test.flatten(), "Prediction":y_pred.flatten()})
print(dt)
```

# Plotting the predicted and actual values

```python
df1=dt.head(10)
df1.plot(kind="bar",figsize=(10,4))
```

# Plotting

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred,color="red")
plt.xlabel("Income")
plt.ylabel("Happiness")
```

# K Nearest Neighbour Algorithm -- Regression

```python
#Importing Libraries
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
#reading and putting data to dataframe
df=pd.read_csv("train.csv")
print(df.head())
```

```python
#We are storing the attributes to X and Y variable by using slicing
X=df[["Height(Inches)"]].values
y=df[["Weight(Pounds)"]].values
```

```python
#Splitting the data into testing and training
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#To find the exact K value
for K in range(1,25):
    model=KNeighborsRegressor(n_neighbors=K,metric="euclidean")
    model.fit(X_train,y_train) #Fit the model
    y_pred=model.predict(X_test) #Make prediction on test set
    error=sqrt(mean_squared_error(y_test,y_pred)) #Calculate rmse
    print("RMSE value for k= ",K,"is:",error)
```

```python
knn=KNeighborsRegressor(n_neighbors=10,metric="euclidean")
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
print(y_pred)
```

```python
error=sqrt(mean_squared_error(y_test,y_pred))
print("RMSE :",error)
```

```python
#Testing the algorithm by giving input values.
testX=[[70.24447]]
print("Prediction :",knn.predict(testX))
```

# SVM in classification

```python
import pandas as pd
from sklearn import metrics
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data)  # 4 cols
df['target'] = iris.target    # adding a col
print(df.head())
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0[0], df0[1],color="green",marker='+')
plt.scatter(df1[0], df1[1],color="blue",marker='*')
```

```python
# **Train Using Support Vector Machine (SVM)**
X = df.iloc[:, :4]
y = df.iloc[:,4]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Making the Confusion Matrix
cm = metrics.confusion_matrix(y_test, y_pred)
print("Confusion MAtrix:\n",cm)
```

```python
#Finding the accuracy of predicted value(Y_pred) by comparing with actual result(y_test).
print('accuracy',metrics.accuracy_score(y_test,y_pred))
#print("Prediction :",model.predict([[4.8,3.0,1.5,0.3]]))
```

```python
# **Tune parameters**
# **1. Regularization (C)**

model_C = SVC(C=1)
model_C.fit(X_train, y_train)
print("Model using c=1 as reg ",model_C.score(X_test, y_test))
```

```python
model_C = SVC(C=10)
model_C.fit(X_train, y_train)
print("Model using c=10 as reg ",model_C.score(X_test, y_test))
```

```python
# **2. Gamma**
# controls the distance of influence of a single training point
# It does not do anything when the kernal is linear in nature
model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)
print("Model score gamma",model_g.score(X_test, y_test))
```

```python
# **3. Kernel**
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
print("Model score Kernal",model_linear_kernal.score(X_test, y_test))
```

# SVM in Regression

```python
# necessary Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Admission_Prediction.csv')
print(df.head())
```

```python
print("NaN Values list before Cleaning ",df.isna().sum())
```

```python
# As we can see, there are some column with missing values. we need to impute those missing values.
df['GRE Score'].fillna(df['GRE Score'].mean(),inplace=True)
df['TOEFL Score'].fillna(df['TOEFL Score'].mean(),inplace=True)
df['University Rating'].fillna(df['University Rating'].mode()[0],inplace=True)
```

```python
# seeing that after imputation no column has missing values
print("NaN Values list after Cleaning ",df.isna().sum())
#print(df['University Rating'].head(10))
```

```python
x=df.drop(['Chance of Admit','Serial No.'],axis=1) # df.iloc[:,1:-1]
y=df['Chance of Admit']
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
#importing the  model
from sklearn.svm import SVR
svr= SVR(C=1)
svr.fit(train_x, train_y)
pred_y=svr.predict(test_x)
```

```python
# Accuracy test
from sklearn.metrics import r2_score
from sklearn import metrics
rmse = np.sqrt(metrics.mean_squared_error(test_y, pred_y))
print("RMSE Score :",rmse)
```

```python
score= r2_score(test_y,svr.predict(test_x))
print("R2 Score : ",score)
```
